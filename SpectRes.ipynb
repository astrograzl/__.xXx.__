```python
import glob
import numpy as np
from sklearn.preprocessing import robust_scale
```

```python
from castor import spec2sqr
from spectres import spectres
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn-darkgrid")
plt.rcParams["figure.figsize"] = (16,9)
plt.rcParams["image.cmap"] = "Spectral"
```

```python
specs = glob.glob("/mnt/arch/POLLUX/spec/*")
spec = np.random.choice(specs)
```

```python
wave, flux, norm = np.loadtxt(spec, unpack=True)
```

```python
new = np.linspace(3690, 9100, 64*64)
lux = spectres(new, wave, flux)
```

```python
plt.subplot(211); plt.plot(wave, flux)
plt.subplot(212); plt.plot(new, lux)
```

```python
plt.imshow(spec2sqr(robust_scale(lux)));
plt.xticks([]); plt.yticks([])
plt.colorbar();
```
