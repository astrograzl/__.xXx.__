````python
import os
import glob
import numpy as np
import pandas as pd
````

````python
from ipywidgets import IntProgress
from IPython.display import display
````

````python
data = pd.read_csv("dr4/catalog/dr4_v2_stellar.csv.gz",
                   sep="|", na_values=[-9999.00, "NULL"],
                   low_memory=False)
````

````python
data.describe()
````

````python
basename = data["lmjd"].map("spec-{:5d}-".format)    \
         + data["planid"].map("{}".format)       \
         + data["spid"].map("_sp{:02d}-".format) \
         + data["fiberid"].map("{:03d}".format)  \
         + ".fits"
data["basename"] = basename
````

```python
DATA = "/space/janak/LAMOST/sample/"
BASE = "http://cdsarc.u-strasbg.fr/ftp/V/153/dr4.lamost.org/sas/fits/"
PRES = glob.glob(os.path.join(DATA, "*/*.fits.gz"))
with open("lamost.url", "w") as file:
    for i in range(len(data)):
        if np.random.randint(101) == 1:
            line = data.iloc[i]
            DIRS = line["planid"]
            FITS = line["basename"]
            NAME = os.path.join(DIRS, FITS)
            #if not NAME in PRES:
            file.write(f"wget -O {DATA+FITS} {BASE+NAME}\n")
```
