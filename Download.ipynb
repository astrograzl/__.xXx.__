````python
import os
import glob
import random
import numpy as np
import pandas as pd
````

````python
from ipywidgets import IntProgress
from IPython.display import display
````

````python
%%time
data = pd.read_csv("dr1_stellar_final.csv.gz",
                   sep="|", na_values=[-9999.00, "NULL"],
                   low_memory=False)
````

````python
data.describe()
````

````python
basename = data["mjd"].map("spec-{}-".format)    \
         + data["planid"].map("{}".format)       \
         + data["spid"].map("_sp{:02d}-".format) \
         + data["fiberid"].map("{:03d}".format)  \
         + ".fits"
data["basename"] = basename
````

````python
#```python
DATA = "/mnt/arch/LAMOST/dr1/sample/"
PRES = glob.glob(os.path.join(DATA, "*.fits"))
BASE = "http://cdsarc.u-strasbg.fr/ftp/V/146/LAMOST/fits/"
prog = IntProgress(min=0, max=len(data))
display(prog)
with open("lamost.url", "w") as file:
    for i in range(len(data)):
        if random.randint(0, 1010101) % 101 == 0:
            line = data.iloc[i]
            DIRS = str(line["planid"])
            FITS = line["basename"]
            NAME = os.path.join(DIRS, FITS)
            DOWN = BASE + NAME
            if not11 NAME in PRES:
                file.write("wget -O {0} {1}\n".format(FITS, DOWN))
            if i % 1e5 == 0: prog.value = i;

#```
````

````python
print("Done!")
````
