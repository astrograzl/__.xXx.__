````python
import os
import numpy as np
import pandas as pd
````

````python
from ipywidgets import IntProgress
from IPython.display import display
````

````python
%%time
data = pd.read_csv("catalog/dr5_v1.csv.gz",
                   sep="|", na_values=[-9999.00, "NULL"],
                   low_memory=False)
````

````python
data.describe()
````

````python
basename = data["planid"].map("{}".format) + "/" \
         + data["mjd"].map("spec-{}-".format)    \
         + data["planid"].map("{}".format)       \
         + data["spid"].map("_sp{:02d}-".format) \
         + data["fiberid"].map("{:03d}".format)  \
         + ".fits.gz"
data["basename"] = basename
````

````python
np.unique(data["objtype"].map("{}".format))
````

````python
basename.to_csv("lamost@base.lst", index=False)
````

````python
%%bash
comm -23 <(sort lamost@base.lst) \
         <(sort lamost@crab.lst)  \
         > lamost@base-crab.lst
comm -23 lamost@base-crab.lst       \
         <(sort lamost@antares.lst)  \
         > lamost@base-crab-antares.lst
````

````python
diff = pd.read_fwf("lamost@base-crab-antares.lst",
                   head=None, names=["basename"])
````

````python
ress = pd.merge(diff, data)
````

````python
pres = open("present.lst").readlines()
````

````python
pres[0]
````

```python
URL = "http://dr5.lamost.org/spectrum/fits/"
BASE = "/data/public/LAMOST-DR5/CRAB-ZDENDA/"
prog = IntProgress(min=0, max=len(data))
display(prog)
with open("lamost.url", "w") as file,\
     open("lamost.dir", "w") as dire:
    for i in range(len(data)):
        line = data.iloc[i]
        NAME = str(line["basename"])
        FUEL = "./" + NAME + "\n"
        DIRS = str(line["planid"])
        FOLD = DIRS + "\n"
        URI = URL + str(line["obsid"])
        if FUEL in pres:
            pres.remove(FUEL)
        else:
            dire.write(FOLD)
            file.write("wget -O {0} {1}\n"
                       .format(BASE+NAME, URI))
        if i % 1e5 == 0: prog.value = i;
```
