```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.style.use("seaborn-darkgrid")
plt.rcParams["font.family"] = "DejaVu Sans"
```

```python
i = 0
a = np.linspace(0, 1, 256*256)\
      .reshape((256, 256))
plt.figure(figsize=(16, 48))
for cmap in plt.cm.cmap_d:
    if cmap[-2:] != "_r":
        plt.subplot(18, 5, 1+i)
        plt.title(cmap)
        plt.grid(False)
        plt.imshow(a, cmap=cmap)
        plt.xticks([]); plt.yticks([])
        i += 1
```
